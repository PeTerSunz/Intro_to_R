## There are 3 things we can do with string

- How long is it?
- Detect
- Extract 
- Replace

```r
## how long i.e. length
str_length(starwars$name)
```

```r
## detect 'Sky' in the name
str_detect(starwars$name, "Sky")
```

```r
## detect - show me the result
str_subset(  tolower(starwars$name) , "sky")
```

```r
## Extract this pattern
str_extract(starwars$name, "Sky")
```

```r
## Extract the first - third character
str_sub(starwars$name, 1, 3)
```

```r
## Replace 'Sky' with 'Thailand'
skywalker_family <- str_subset(starwars$name, "Sky") 

print(str_replace(skywalker_family, "Sky", "Thailand"))
```

```r
result <- str_split(starwars$name, " ", simplify=T)
```
